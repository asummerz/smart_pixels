```python
from PIL import Image
from pathlib import Path
import pandas as pd
import os
import random
import json
```

```python
train_raw = Path("/mnt/d//Data/119.IR 실제 환경 안면 검출 및 인식 데이터/01-1.정식개방데이터/Validation/01.원천데이터/VS/01.Recognition")
train_label = Path("/mnt/d/Data/119.IR 실제 환경 안면 검출 및 인식 데이터/01-1.정식개방데이터/Validation/02.라벨링데이터/VL/01.Recognition")
```

```python
# 랜덤하게 선택할 파일의 개수
num_files_to_select = 20

# 선택된 파일들을 저장할 리스트 초기화
selected_files = []
files_path = []

for root, dirs, files in os.walk(train_raw):
    # 랜덤하게 파일 선택
    selected_files.extend([os.path.join(root, file) for file in random.sample(files, min(num_files_to_select, len(files)))])
```

```python
df_train = pd.DataFrame(selected_files, columns=["filepath"])
```

```python
label_train = []
for root, dirs, files in os.walk(train_label):
    for file in files:
        filepath = os.path.join(root, file)
        label_train.append(filepath)
```

```python
df_label_path = pd.DataFrame(label_train, columns=["label_path"])
df_label_path.head()
```

```python
label_data = []
for index, row in df_label_path.iterrows():
    filepath = row["label_path"]
    if os.path.exists(filepath):  # 파일이 존재하는지 확인
        with open(filepath, 'r') as file:
            file_content = json.load(file)
            label_data.append(file_content)
    else:
        print(f"파일 '{filepath}'이(가) 존재하지 않습니다.")
```

```python
label_data
```

```python
df_label = pd.json_normalize(label_data)
df_label
```

```python
df_label_merge = pd.merge(df_train, df_label, how="inner", on="filename")
df_label_merge
```

```python
df_label_merge.to_csv("./merged_label.csv", index=False)
```
